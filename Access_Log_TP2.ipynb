##QUESTION 1

Effectuer un  MAP/REDUCER, permettant de trouver le nombre fois qu'un chemin a été accessible

**MAP**

```python
%%writefile mapper.py
#!/usr/bin/python
import sys
for line in sys.stdin:
  data = line.strip().split(" ")
  if len(data) == 10:
    addrIP, identity, nom_user, heure_fin, zone, methode, chemin, chaine, code_etat, taille = data
    print "{0}\t{1}".format(chemin, code_etat)
```

```python
!head -15 ../content/TP/data/access_log | ./mapper.py
```

**REDUCER**

```python
%%writefile reducer.py
#!/usr/bin/python
import sys
visite_Total = 0
oldAddr = None

for line in sys.stdin:
  data = line.strip().split("\t")
  if len(data) != 2:
    continue
  chemin, code_etat= data 
  if oldAddr and oldAddr != chemin:
    print("{0}\t{1}").format(oldAddr, visite_Total)
    visite_Total = 0

  oldAddr = chemin
  if int(code_etat) != 404: 
    visite_Total += 1

if oldAddr != None:
  print(oldAddr,"\t", visite_Total)
```

```python
## Ouverture du shelle pour changer les droits d'usage (+x) sur le fichier mapper
#le rendre executable
getshell(height=100)
#getshell(height=400) # custom height of the terminal
```

```python
!head -15 ../content/TP/data/access_log |./mapper.py |sort | ./reducer.py
```

**Execution du Framework** en streaming pour calculer toute la tache sur tout le dataset
Le resultat sera recupere dans joboutput

```python
!/usr/local/hadoop-3.3.3/bin/hadoop jar /usr/local/hadoop-3.3.3/share/hadoop/tools/lib/hadoop-streaming-3.3.3.jar -input /content/TP/data/access_log -output /content/TP/code/joboutput -file mapper.py  -file reducer.py  -mapper mapper.py  -reducer reducer.py
```
